```python
import pandas as pd


# List of columns to convert to numeric types
columns_to_convert = [
    'OPEN', 'HIGH', 'LOW', 'PREV._CLOSE', 'LTP',
    'CHNG', 'VOLUME_shares', 'VALUE_cr',
    '52W_H', '52W_L', '30D_%CHNG', '365D_%CHNG'
]

# Convert the columns to numeric types
for column in columns_to_convert:
    df[column] = pd.to_numeric(df[column].str.replace(',', ''), errors='coerce')

# Display basic statistics after conversion
print(df.info())
print(df.describe(include='all'))
```

```python
df.info()
```

```python
import pandas as pd



# Ensure the OPEN column is numeric
df['OPEN'] = pd.to_numeric(df['OPEN'], errors='coerce')

# Sort the dataframe by the OPEN column in descending order
sorted_df = df.sort_values(by='OPEN', ascending=False)

# Select the top 10 stocks with the highest opening prices
top_10_stocks = sorted_df.head(10)

print(top_10_stocks[['SYMBOL', 'OPEN']])
```

```python
# Ensure the LTP column is numeric
df['LTP'] = pd.to_numeric(df['LTP'], errors='coerce')

# Sort the dataframe by the LTP column in ascending order (lowest to highest)
sorted_df = df.sort_values(by='LTP', ascending=True)

# Select the top 10 stocks with the lowest closing prices
top_10_stocks = sorted_df.head(10)

print(top_10_stocks[['SYMBOL', 'LTP']])
```

```python
# Ensure the %CHNG column is numeric
df['%CHNG'] = pd.to_numeric(df['%CHNG'], errors='coerce')

# Sort the dataframe by the %CHNG column in descending order
sorted_df = df.sort_values(by='%CHNG', ascending=False)

# Select the top stocks with the highest percentage change
top_stocks = sorted_df.head()

print(top_stocks[['SYMBOL', '%CHNG']])
```

```python
# Ensure the 30D_%CHNG column is numeric
df['30D_%CHNG'] = pd.to_numeric(df['30D_%CHNG'], errors='coerce')

# Sort the dataframe by the 30D_%CHNG column in descending order
sorted_df = df.sort_values(by='30D_%CHNG', ascending=False)

# Select the top stocks with the highest 30-day percentage change
top_stocks = sorted_df.head()

print(top_stocks[['SYMBOL', '30D_%CHNG']])
```

```python
# Calculate the difference between high and low prices
df['High_Low_Difference'] = df['HIGH'] - df['LOW']

# Sort the DataFrame by the difference between high and low prices in descending order
top_difference_stocks = df.sort_values(by='High_Low_Difference', ascending=False).head(10)

print("Top 10 stocks with the largest difference between high and low prices in a single day:")
print(top_difference_stocks[['SYMBOL', 'High_Low_Difference']])
```

```python
# Define threshold values for capping outliers
percentage_change_threshold = 3  # Example threshold for percentage change (%CHNG)
volume_threshold = 1e8  # Example threshold for trading volume (VOLUME_shares)

# Cap outliers for percentage change (%CHNG)
df['%CHNG_capped'] = df['%CHNG'].clip(lower=-percentage_change_threshold, upper=percentage_change_threshold)

# Cap outliers for trading volume (VOLUME_shares)
df['VOLUME_shares_capped'] = df['VOLUME_shares'].clip(upper=volume_threshold)

# Display the capped columns
print("Capped Percentage Change (%CHNG):")
print(df['%CHNG_capped'].describe())
print("\nCapped Trading Volume (VOLUME_shares):")
print(df['VOLUME_shares_capped'].describe())
```
